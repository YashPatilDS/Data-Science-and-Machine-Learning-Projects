```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
fdi_df = pd.read_csv("C:/Unified Mentor Data Science Projects/FDI data.csv")
fdi_df.head()
```

```python
fdi_df.set_index('Sector', inplace=True)
```

```python
fdi_df.head()
```

```python
total_investment_per_sector = fdi_df.sum(axis=1)
total_investment_per_sector
```

```python
annual_investment_trend = fdi_df.sum(axis=0)
annual_investment_trend
```

```python
plt.figure(figsize=(10, 8))
for sector in fdi_df.index:
    plt.plot(fdi_df.columns, fdi_df.loc[sector], label=sector)
plt.title('Sector-wise Investment Trend: 2000-01 - 2016-17')
plt.xlabel('Financial Year')
plt.ylabel('FDI in Million USD')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
top_sectors_by_fdi = total_investment_per_sector.sort_values(ascending=False).head(10)
print(top_sectors_by_fdi)
```

```python
correlation_matrix = fdi_df.transpose().corr()
plt.figure(figsize=(8, 8))
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix of FDI between Sectors')
plt.show()
```

```python
plt.figure(figsize=(7, 8))
top_sectors_by_fdi.sort_values(ascending=False).plot(kind='bar')
plt.title('Top sectors by FDI')
plt.xlabel('Sector')
plt.ylabel('Total FDI in Million USD')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.pie(top_sectors_by_fdi, labels=top_sectors_by_fdi.index, autopct='%1.1f%%')
plt.title('Top Sectors by Total FDI (in Million USD)')
plt.show()
```

```python
plt.figure(figsize=(12, 14))
total_investment_per_sector.sort_values(ascending=False).plot(kind='bar')
plt.title('Total Investment per Sector (2000-01 to 2016-17)')
plt.xlabel('Sector')
plt.ylabel('Total FDI in Million USD')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
annual_investment_trend.plot(kind='line', marker='o')
plt.title('Annual Investment Trend: 2000-01 - 2016-17')
plt.xlabel('Financial Year')
plt.ylabel('Total FDI in Million USD')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
for sector in top_sectors_by_fdi.index:
    plt.plot(fdi_df.columns, fdi_df.loc[sector], marker='o', label=sector)
plt.title('Detailed Sector-wise Investment Trend for Top Sectors (2000-01 to 2016-17)')
plt.xlabel('Financial Year')
plt.ylabel('FDI in Million USD')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

```python
# Calculating and Printing KPIs
total_fdi = total_investment_per_sector.sum()
average_annual_investment = annual_investment_trend.mean()

max_investment_value = annual_investment_trend.max()
max_investment_year = annual_investment_trend.idxmax()

min_investment_value = annual_investment_trend.min()
min_investment_year = annual_investment_trend.idxmin()

ighest_fdi_value = total_investment_per_sector.max()
highest_fdi_sector = total_investment_per_sector.idxmax()  # Define the highest FDI sector

lowest_fdi_value = total_investment_per_sector.min()
lowest_fdi_sector = total_investment_per_sector.idxmin()

```

```python
print("Key Performance Indicators (KPIs):")

print(f"Total FDI (2000-01 to 2016-17): {total_fdi:.2f} Million USD")

print(f"Average Annual Investment: {average_annual_investment:.2f} Million USD")

print(f"Maximum Annual Investment: {max_investment_year} ({max_investment_value:.2f} Million USD)")

print(f"Minimum Annual Investment: {min_investment_year} ({min_investment_value:.2f} Million USD)")

print(f"Sector with Highest Total FDI: {highest_fdi_sector} ({highest_fdi_value:.2f} Million USD)")

print(f"Sector with Lowest Total FDI: {lowest_fdi_sector} ({lowest_fdi_value:.2f} Million USD)")
```
